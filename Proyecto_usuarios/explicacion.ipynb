## Variable archivo

la variable ARCHIVO almacena el nombre del archuivo donde se gurdaran  los archivos registrados, se utiliza para evitar escribirvarias veces el nombre del archivo dentro del programa y facilitar futuros cambios.

La variable ARCHIVO almacena el nombre del archivo donde se guardará la información de los usuarios.

ARCHIVO = "usuarios.txt"

El valor "usuarios.txt" es un dato de tipo str (string), ya que se encuentra entre comillas y representa texto.

Se utiliza una constante para evitar escribir varias veces el nombre del archivo dentro del programa y facilitar cambios futuros en caso de que el archivo tenga otro nombre.

## Funciones

Las funciones permiten agrupar instrucciones para realizar una tarea específica dentro del programa.

Se crean utilizando la palabra reservada def.

En este proyecto se utiliza la función registrar_usuario() para organizar todo el proceso de registro de usuarios en un solo bloque de código, facilitando la lectura y el mantenimiento del programa.

## Uso de try y except

La estructura try-except se utiliza para manejar errores durante la ejecución del programa.

El bloque try contiene instrucciones que pueden generar errores, mientras que except permite capturar esos errores y mostrar mensajes amigables al usuario sin detener el programa.

En este proyecto se utiliza para evitar errores cuando el usuario ingresa datos incorrectos, especialmente al convertir la edad a un número entero.

Las funciones ayudan a evitar repetir código y permiten reutilizar instrucciones cuando sea necesario.

### Error ValueError

El error `ValueError` ocurre cuando se intenta convertir un dato a otro tipo y la conversión no es posible.

En este proyecto puede ocurrir cuando el usuario ingresa texto en lugar de un número para la edad.

Por ejemplo:

```python
int("hola")
```

En este caso Python genera un `ValueError`, el cual es capturado mediante `except ValueError` para mostrar un mensaje amigable al usuario.

with open(ARCHIVO, "a", encoding="utf-8") as archivo:

Esto es útil porque cada vez que se registra un nuevo usuario, la información se añade al archivo conservando los registros anteriores.
